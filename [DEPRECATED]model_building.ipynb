```python
place_details = []
for i in range(len(restaurant)):
    url = f"{detailURL}{restaurant.lat.iloc[i]},{restaurant.long.iloc[i]}{detailURL2}{key}"
    print(url)
    if requests.get(url).json()["status"] == "OK":       
        try:
            place_details.append({
                "lat":requests.get(url).json()["results"][0]["geometry"]["location"]["lat"],
                "long":requests.get(url).json()["results"][0]["geometry"]["location"]["lng"],
                "rating":requests.get(url).json()["results"][0]["rating"],
                "total_review":requests.get(url).json()["results"][0]["user_ratings_total"],
                "name":requests.get(url).json()["results"][0]["name"]
            })
        except (KeyError, IndexError) as e:
            try:
                place_details.append({
                    "lat":requests.get(url).json()["results"][1]["geometry"]["location"]["lat"],
                    "long":requests.get(url).json()["results"][1]["geometry"]["location"]["lng"],
                    "rating":requests.get(url).json()["results"][1]["rating"],
                    "total_review":requests.get(url).json()["results"][1]["user_ratings_total"],
                    "name":requests.get(url).json()["results"][1]["name"]
                })
            except (KeyError, IndexError) as e:
                try:
                    place_details.append({
                        "lat":requests.get(url).json()["results"][2]["geometry"]["location"]["lat"],
                        "long":requests.get(url).json()["results"][2]["geometry"]["location"]["lng"],
                        "rating":None,
                        "total_review":None,
                        "name":requests.get(url).json()["results"][2]["name"]
                    })
                except IndexError:
                    place_details.append({
                        None
                    })
```
